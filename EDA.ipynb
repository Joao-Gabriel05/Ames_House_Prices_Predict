```python
import pandas as pd

def compute_coefficient_of_variation(data: pd.DataFrame) -> pd.Series:
    """
    Computes the coefficient of variation for each numeric column in a DataFrame.
    """
    # note os parênteses que permitem quebrar a linha sem backslashes
    stats = (
        data
        .select_dtypes(include='number')
        .describe()
        .transpose()
    )
    cv = stats['std'] / stats['mean']
    cv.rename('Coefficient of Variation', inplace=True)
    return cv

# uso:
cv_series = compute_coefficient_of_variation(data_15).round(2)
cv_sorted = cv_series.sort_values(ascending=False)
cv_df = cv_sorted.reset_index()
cv_df.columns = ['Feature', 'Coefficient of Variation']
print(cv_df.to_markdown(index=False))

```

- Para padronizar as categorias, defini um limiar de 5 %: todas as classes com participação inferior a esse valor serão reunidas na categoria `Other`, preservando a maior parte da informação original e reduzindo o risco de overfitting. Para variáveis ordinais — por exemplo, `Overall.Qual` — essa abordagem não se aplica bem. Nesses casos, ainda usando o limiar de 5 %, agruparei as classes em faixas, gerando categorias do tipo `Less than X` e `More than X`, de forma a manter a ordenação natural e controlar a granularidade.
- Para `Neighborhood`, a coluna `Other` pode acabar ficando com um proporção muito grande, isso é porque tem muitas categorias nessa feature, portanto nosso limiar deve ser menor

```python
data_15['Neighborhood'].value_counts(normalize=True)
```

Como observado, ao definir um limiar de 5%, a categoria `Other` acaba concentrando uma fatia muito grande do conjunto — e mesmo reduzindo esse limiar, `Other` continua volumosa em razão do alto número de bairros com representações mínimas. Portanto, para este caso, optarei por descartar (drop) todos os bairros cuja participação seja inferior a 1 %.

```python

threshold = 0.01

neigh_prop = data_15['Neighborhood'].value_counts(normalize=True)

drop_neigh = neigh_prop[neigh_prop < threshold].index.tolist()
print(f'Dropping {len(drop_neigh)} neighborhoods (proportion < {threshold*100:.0f}%):')
print(drop_neigh)

data_15_filtered = data_15[~data_15['Neighborhood'].isin(drop_neigh)].copy()

print("\nProporções após o drop:")
print(data_15_filtered['Neighborhood'].value_counts(normalize=True).head(10))

original_count = len(data_15)
filtered_count = len(data_15_filtered)
dropped_count = original_count - filtered_count

dropped_pct = dropped_count / original_count * 100

print(f"{dropped_count} linhas removidas de {original_count} ({dropped_pct:.2f}% do dataset).")
```

Portanto, com essa remoção perdemos 3.28% de informação do nosso dataset. Acredito que apesar da perda de informação, limitar o número de bairros pode ser algo benefico para o nosso modelo

```python
def group_rare_categories_auto(data, column, threshold=0.05, new_label='Other'):
    """
    Agrupa automaticamente categorias que aparecem menos que o threshold em 'Other'.
    """
    data = data.copy()
    proportions = data[column].value_counts(normalize=True)
    rare_labels = proportions[proportions < threshold].index.tolist()
    
    data[column] = data[column].apply(lambda x: new_label if x in rare_labels else x)
    
    return data
```

```python
import pandas as pd
from typing import List, Optional

def remove_outliers_iqr(
    df: pd.DataFrame,
    columns: Optional[List[str]] = None,
    multiplier: float = 1.5
) -> pd.DataFrame:
    """
    Remove outliers de um DataFrame usando o método IQR,
    imprimindo os limites e a quantidade de dados perdidos.

    Args:
        df: DataFrame de entrada.
        columns: lista de colunas numéricas nas quais aplicar a remoção.
                 Se None, usa todas as colunas numéricas do df.
        multiplier: fator para definir os limites (padrão 1.5).

    Returns:
        Um novo DataFrame sem as linhas que contenham outliers em qualquer
        das colunas especificadas.
    """
    original_count = len(df)

    if columns is None:
        columns = df.select_dtypes(include='number').columns.tolist()

    mask = pd.Series(True, index=df.index)

    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - multiplier * IQR
        upper_bound = Q3 + multiplier * IQR

        col_mask = df[col].between(lower_bound, upper_bound)
        removed_by_col = (~col_mask).sum()

        print(
            f"Coluna '{col}':\n"
            f"  Q1 = {Q1:.3f}, Q3 = {Q3:.3f}, IQR = {IQR:.3f}\n"
            f"  lower_bound = {lower_bound:.3f}, upper_bound = {upper_bound:.3f}\n"
            f"  linhas removidas nesta coluna = {removed_by_col}\n"
        )


        mask &= col_mask

    filtered_df = df.loc[mask].copy()
    removed_total = original_count - len(filtered_df)
    pct_removed = removed_total / original_count * 100

    print(
        f"Total de linhas originais = {original_count}\n"
        f"Total de linhas após remoção = {len(filtered_df)}\n"
        f"Total de linhas removidas = {removed_total} ({pct_removed:.2f}% do dataset)\n"
    )

    return filtered_df,lower_bound, upper_bound


```

```python

from typing import List, Optional

def remove_outliers_iqr(
    df: pd.DataFrame,
    columns: Optional[List[str]] = None,
    multiplier: float = 1.5
) -> pd.DataFrame:
    """
    Remove outliers de um DataFrame usando o método IQR,
    imprimindo os limites e a quantidade de dados perdidos.

    Args:
        df: DataFrame de entrada.
        columns: lista de colunas numéricas nas quais aplicar a remoção.
                 Se None, usa todas as colunas numéricas do df.
        multiplier: fator para definir os limites (padrão 1.5).

    Returns:
        Um novo DataFrame sem as linhas que contenham outliers em qualquer
        das colunas especificadas.
    """
    original_count = len(df)

    if columns is None:
        columns = df.select_dtypes(include='number').columns.tolist()

    mask = pd.Series(True, index=df.index)

    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - multiplier * IQR
        upper_bound = Q3 + multiplier * IQR

        col_mask = df[col].between(lower_bound, upper_bound)
        removed_by_col = (~col_mask).sum()


        mask &= col_mask

    filtered_df = df.loc[mask].copy()
    removed_total = original_count - len(filtered_df)
    pct_removed = removed_total / original_count * 100



    return filtered_df


```
